### What is the mean age of users?

```python
# your code here
print(df.age.mean())
```

### What is the mean age per occupation

```python
# your code here
print(df.groupby(by='occupation').mean('age').age)
```

### For each occupation, calculate the minimum and maximum ages

```python
# your code here
print('Maximum ages:')
print(df.groupby(by='occupation').max('age').age)
print('Minimum ages:')
print(df.groupby(by='occupation').min('age').age)
```

### For each combination of occupation and gender, calculate the mean age

```python
# your code here
print(df.groupby(by=['occupation','gender']).mean('age').age)
```
